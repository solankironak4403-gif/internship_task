```python
#numaric on Pclass
df['Pclass'].describe()
```

```python
df['Pclass'].isnull().sum()
```

```python
count = df['Pclass'].value_counts()
plt.bar(count.index, count)
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Pclass Value Counts')
plt.show()
```

```python
count = df['Pclass'].value_counts()
plt.pie(
    count,
    labels= count.index,
    autopct='%1.1f%%',
    startangle=360
)
plt.title('Pclass Value Counts')
plt.show()
```

```python
# numaric on SibSp
df['SibSp'].describe()
```

```python
df['SibSp'].isnull().sum()
```

```python
counts = df['SibSp'].value_counts().sort_index()

plt.bar(counts.index.astype(str), counts.values)
plt.xlabel('Number of Siblings / Spouses (SibSp)')
plt.ylabel('Count')
plt.title('Distribution of SibSp')
plt.show()
```
