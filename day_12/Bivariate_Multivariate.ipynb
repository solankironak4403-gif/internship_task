### Bivariate Analysis

### Numerical vs Numerical

```python
#correlation
df[['Salary', 'PerformanceScore']].corr()
```

### Visual Analysis

```python
### scatter plot 
plt.figure(figsize=(7,8))
sns.scatterplot(x='Age',y='Salary',data=df)
plt.title('Age vs Salary Sactter plot')
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.scatterplot(x='Experience',y='Salary',data=df)
plt.title('Experience Vs Salary Scatter plot')
plt.show()
```

### Numerical vs Categorical

```python
#Groupby Department and calculate the avg salary
df.groupby('Department')['Salary'].mean().reset_index().sort_values(by='Salary', ascending=False)
```

```python
# box plot of gender vs salary
plt.figure(figsize=(7,6))
sns.boxplot(x='Gender',y='Salary',data=df,palette='Set2',hue='Gender')
plt.title('Boxplot of Gender Vs Salary')
plt.show()
```

```python
# Boxplot Of Salary Vs Department
plt.figure()
sns.boxplot(x='Department', y='Salary', data=df,hue='Department',palette='Set2')
plt.title("Salary Distribution by Department")
plt.show()
```

```python
#Average Salary per Department
plt.figure(figsize=(6,7))
sns.barplot(x='Department',y='Salary',data=df,estimator='mean',palette='Set3',hue='Department')
plt.title('Average Salary per Department')
```

### Multivariate Analysis

```python
#pairplot of Columns 
sns.pairplot(df[['Age','Salary','Experience','PerformanceScore']])
plt.show()
```

```python
#Correlation Matrix Or Heatmap
plt.figure(figsize=(8,6))
corr = df[['Age','Salary','Experience','PerformanceScore']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#insight
# The bivariate and multivariate analysis shows that experience has a strong positive relationship with salary.
# Age has a moderate impact on salary but is not a strong factor for performance.
# Management and Finance departments offer higher salaries compared to IT and HR.
# Salary distribution across gender shows no major bias in the dataset.
# Pair plots and correlation matrix confirm that experience is the key factor influencing both salary and performance.
```
