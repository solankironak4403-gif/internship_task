### 1) Categorical Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
#categorial on  Embarked
df['Embarked'].value_counts()
```

```python
df['Embarked'].value_counts(normalize=True) * 100
```

```python
df['Embarked'].isnull().sum()
```

```python
count = df['Embarked'].value_counts()
Labels = ['Southampton', 'Cherbourg', 'Queenstown']
plt.bar(Labels, count)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Embarked Value Counts')
plt.show()
```

```python
count = df['Embarked'].value_counts()
plt.pie(
    count,
    labels = count.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Embarked Value Counts')
plt.show()
```

```python
# Categorial on sex
df['Sex'].value_counts()
```

```python
df['Sex'].value_counts(normalize=True) * 100
```

```python
count = df['Sex'].value_counts()
plt.bar(count.index, count)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Value Counts')
plt.show()
```

```python
plt.pie(
    count,
    labels = count.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Sex Value Counts')
plt.show()
```
