```python
df = pd.read_csv('emp.csv')
df.head()
```

```python
df.describe()
```

```python
# cheacking missing values
df.isnull().sum()
```

```python
# handling missing values by filling with mean for salary and age and median for joindate
df.fillna(
    {
        'Age': df['Age'].mean(),
        'Salary': df['Salary'].mean()
    },
    inplace=True
)
```

```python
#check the duplicate values
df.duplicated().sum()
```

```python
#Data Type Conversion
df['JoiningDate'] = pd.to_datetime(df['JoiningDate'], errors='coerce')
```

```python
df.fillna(
    {
        'JoiningDate': df['JoiningDate'].median()
    },
    inplace=True
)
```
